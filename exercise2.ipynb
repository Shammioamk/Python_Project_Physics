```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#import the data
df = pd.read_csv('Raw Data.csv')
df.head()
#df.info()
```

```python
#To access only single column:
#df['Time (s)'] 


# Let's make a simply graphics
plt.figure(figsize=(15,8))

plt.subplot(3,1,1)
plt.plot(df['Time (s)'],df['Linear Acceleration x (m/s^2)'])
plt.grid()
plt.title('Acceleration x')
plt.xlabel('Time')
plt.ylabel('a_x')
plt.axis([0, 30, -10, 10]) 

plt.subplot(3,1,2)
plt.plot(df['Time (s)'],df['Linear Acceleration y (m/s^2)'])
plt.grid()
plt.title('Acceleration y')
plt.xlabel('Time')
plt.ylabel('a_y')
plt.axis([0, 30, -10, 10]) 

plt.subplot(3,1,3)
plt.plot(df['Time (s)'],df['Linear Acceleration z (m/s^2)'])
plt.grid()
plt.title('Acceleration z')
plt.xlabel('Time')
plt.ylabel('a_z')
plt.axis([0, 30, -10, 10]) 
plt.show()
```
