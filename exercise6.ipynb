```python

#Import libraries
import numpy as np
import matplotlib.pyplot as plt

#Libraries for analytical solutions (Solve equation and get equation, pen and paper style)
#Symbolic Python
import sympy as sm  #pip install sympy (if needed)

#Numerical solution (define the equation in terms of numbers and get the result as numbers)
#Scientific Python
import scipy as sc #pip install scipy (if needed)
```

Differential Equations

$y' = 5e^{-x} -2y$


```python
#Analytical solution
from sympy.abc import x

y = sm.Function('y') #Define that y is a function
#Write the equation in to the form where we have only zero at the other side
#y' -5*e^-x +2y =0

#Write the terms in a different foemat
#y' --> sm.Derivative(y(x),x)
#- 5*e^-x --> -5*sm.exp(-x)
#2y --> 2*y(x)

result = sm.dsolve(sm.Derivative(y(x),x) -5*sm.exp(-x) +2*y(x),y(x))

result
```

```python
#Numerical solution

#Define the differential equation as a Python function
#Must be in the form y' = 5*e^-x - 2y

def diffun(y,x):
    dydx = 5*np.exp(-x) - 2*y
    return dydx

#Define values of x (range where we define the solution)
x = np.linspace(0,10,101)

#Boundary condition
y0 = 1 # y=1 when x = 0 --> y(0) = 1

y = sc.integrate.odeint(diffun,y0,x)
#y

#From analytical solution, using the boundary condition y(0) = 1
#--> (C + 5)= 1 --> C = -4

plt.plot(x,y,'*')
plt.xlabel('x')
plt.ylabel('y')
#Add analytical solution 
C = -4
plt.plot(x, (C*np.exp(-x)+5)*np.exp(-x))
plt.legend(['Numerical solution','Analytical solution'])
```

Exercise 4a

y' +7y = 0

```python
#Analytical solution, using scipy.dsolve function

#dsolve requires that we write the equation in format where we have only zero on other side
#Now we already have this format
#We must write each term in a correct way

#y' --> sm.Derivative(y(x),x) #Compare to dy(x)/dx
#7y --> 7*y(x)
from sympy.abc import x #x is variable
y = sm.Function('y')    #y is a function

solution = sm.dsolve(sm.Derivative(y(x),x) + 7*y(x),y(x))
#solution = sm.dsolve(y'+7y, y)
#Inside dsolve define the differential equation and what are we solving
solution
```

```python
#Numerical solution 
#This important because many differential equations don't have the analytical solution

#Let's write the equation in the different format
#y' + 7y = 0 --> dydx + 7*y = 0 --> dydx = - 7*y

#Write the differential equation as a Python function
#Inputs are y and x, returns y'

def f11a(y,x):
     dydx = - 7*y #Same as y' + 7y = 0 
     return dydx

#Define the range where we look for a solution
X = np.linspace(0,10,101) #X is an array of values from 0 to 10 between 0.1

#Boundary condition (numerical solution cannot include C)
y0 = 2 #Defines that the value of y is 2 in the first element of X --> y(0) = 2

num_solution = sc.integrate.odeint(f11a,y0,X) #Equation, boundary condition, X values

#Is this the curve y(x) = C*e^(-7x) with some value of C? Let's investigate:
#Boundary condition y(0) = 2 --> y(0) = C --> C = 2
#-- > y(x) = 2*e^(-7x)
plt.figure(figsize=(7,5))
plt.plot(X,num_solution,'*') #Numerical solution
C = 2
plt.plot(X,C*np.exp(-7*X)) #Analytical solution
plt.legend(['Numerical','Analytical'])
plt.xlabel('x')
plt.ylabel('y(x)')
plt.grid()
plt.show()
```

Task 4b

y' + cos(x)y = 0

```python
 #Analytical solution
 #y' + cos(x)y = 0

#Terms in the correct format
#y' --> sm.Derivative(y(x),x)
#cos(x)y --> sm.cos(x)*y(x)
from sympy.abc import x #x is variable
y = sm.Function('y')    #y is a function

solution = sm.dsolve(sm.Derivative(y(x),x) + sm.cos(x)*y(x),y(x))
solution
```

```python
#Numerical solution

#Define the function
#y' + cos(x)y = 0 --> dydx = -np.cos(x)*y

def f4b(y,x):
    dydx = -np.cos(x)*y
    return dydx

#Define the range where we look for a solution
X = np.linspace(0,10,101) #X is an array of values from 0 to 10 between 0.1

#Boundary condition (numerical solution cannot include C)
y0 = 2 #Defines that the value of y is 2 in the first element of X --> y(0) = 2

num_solution = sc.integrate.odeint(f11b,y0,X) #Equation, boundary condition, X values

plt.figure(figsize=(7,5))
plt.plot(X,num_solution) #Numerical solution"
   
```

xy' + 3x^2 * y = 0

```python
#Analytical solution

#Let's write the terms in the proper format
#xy'    --> x * sm.Derivative(y(x),x)
#3x^2 y --> 3*x**2*y(x)
from sympy.abc import x #x is variable
y = sm.Function('y')    #y is a function

solution = sm.dsolve(x * sm.Derivative(y(x),x) + 3*x**2*y(x),y(x))
solution
   
```

```python
#Define the function
# xy' + 3x^2 * y = 0 (divide by x)
#--> y' + 3xy = 0 --> y' = -3yx
#dydx = -3*y*x

def f4c(y,x):
    dydx = -3*y*x
    return dydx
########### copy-paste EXCEPT function name ############

#Define the range where we look for a solution
X = np.linspace(0,10,101) #X is an array of values from 0 to 10 between 0.1
#Boundary condition (numerical solution cannot include C)
y0 = 2 #Defines that the value of y is 2 in the first element of X --> y(0) = 2

num_solution = sc.integrate.odeint(f11c,y0,X) #Equation, boundary condition, X values

plt.figure(figsize=(7,5))
plt.plot(X,num_solution,'*') #Numerical solution
C = 2
plt.plot(X,C*np.exp(-3/2*X**2)) #Analytical solution
plt.legend(['Num','Analog'])
```
